```python
import pandas as pd

def encontrar_categorias(df_title_basics, num_categorias):
    
    filmes = df_title_basics[df_title_basics['titleType'].isin(['movie', 'short'])]
    
    categorias_mais_comuns = df_title_basics.loc[filmes.index, 'genres'].str.split(',').explode().value_counts()
    
    return categorias_mais_comuns.index[:num_categorias]

num_categorias = 10

categorias_mais_comuns = encontrar_categorias(df_title_basics, num_categorias)

print('As categorias de filmes mais comuns são:')

resultado = pd.DataFrame({'Categorias': categorias_mais_comuns})

display(resultado)

print('-=' * 50)
```

```python
import pandas as pd

def mediana_estreia(df_title_basics, df_title_ratings):

    
    df_title_basics['startYear'] = df_title_basics['startYear'].fillna(0)

    
    df_title_basics['startYear'] = df_title_basics['startYear'].astype(int)

    
    df_mesclar = pd.merge(df_title_basics, df_title_ratings, on='tconst', how='inner')

    df_filmes = df_mesclar[(df_mesclar['titleType'].isin(['movie', 'short']))]

    df_mediana = df_filmes.groupby(['originalTitle', 'startYear'], as_index=False)['averageRating'].median()
   
    df_mediana['averageRating'] = df_mediana['averageRating'].apply(lambda x: round(x, 2))

    df_mediana = df_mediana[df_mediana['originalTitle'].str.len() > 3]

    return df_mediana.head(20)

mediana_estreia = mediana_estreia(df_title_basics, df_title_ratings)

print("Mediana de Avaliação dos Filmes em relação ao ano de estreia:")

display(mediana_estreia)

print('-='*50)
```

```python
import pandas as pd
import numpy as np

def duracao_genero(df_title_basics):
    
    df_title_basics['runtimeMinutes'] = pd.to_numeric(df_title_basics['runtimeMinutes'], errors='coerce').fillna(0)
    
    df_title_basics['genres'].fillna('Desconhecido', inplace=True)
    
    media_duracao = df_title_basics.groupby('genres')['runtimeMinutes'].mean().reset_index()
    
    media_duracao['horas'] = media_duracao['runtimeMinutes'] // 60
    media_duracao['minutos'] = media_duracao['runtimeMinutes'] % 60
    media_duracao['segundos'] = 0
    
    media_duracao['duracao_genres'] = media_duracao[['horas', 'minutos', 'segundos']].apply(lambda x: '{:02d}:{:02d}:{:02d}'.format(int(x.iloc[0]), int(x.iloc[1]), int(x.iloc[2])), axis=1)
    
    return media_duracao[['genres', 'duracao_genres']]


media_duracao = duracao_genero(df_title_basics)

print('A relação entre duração e gênero é demonstrada pelo tempo de cada gênero:')

display((media_duracao).head(15))

print('-='*50)
```

```python
import pandas as pd

chunk_size = 100000

chunks = []

try:
    for chunk in pd.read_csv(r"C:\Users\Thiago10\OneDrive\Área de Trabalho\ARQ_JUMP_START\CASEPYTHON\title.ratings.tsv\title.ratings.tsv", sep='\t', na_values='\\N', chunksize=chunk_size, low_memory=False):
        chunks.append(chunk)

    df_title_ratings = pd.concat(chunks, ignore_index=True)

except pd.errors.ParserError as e:
    print(f"Erro na leitura do arquivo: {e}")

df_title_ratings.head() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def catalogo_filmes(df_title_basics, df_title_ratings):
    if 'tconst' not in df_title_basics.columns or 'tconst' not in df_title_ratings.columns:
        print("Erro: O Código do filme não foi encontrado nos índices.")
        return None

    df_filmes = pd.merge(df_title_basics, df_title_ratings, on='tconst', how='inner')
    filmes_recentes = df_filmes[df_filmes['startYear'].fillna(0).astype(int) > 2017]
    
    if len(filmes_recentes) < 20:
        print("Ajuste a pesquisa. Por favor, a quantidade de filmes desejada.")
        return None

    filmes_aleatorios = filmes_recentes.sample(20, replace=False)

    catalogo = {}
    for idx, filme in filmes_aleatorios.iterrows():
        catalogo[idx] = {
            "titulo": filme['primaryTitle'],
            "ano": str(int(filme['startYear'])),
            "genero": filme['genres'] if pd.notna(filme['genres']) else '',
            "preco": 5.00,
            "avaliacao": filme['averageRating'],
            "diaria": 5.00
        }

    return catalogo

def locacao_filmes(catalogo):
    if catalogo is None:
        return 0, {}, []

    total_locacao = 0.0
    filmes_locados = {}
    avaliacoes = {}
    generos = []

    print("\n----------------------------------------------------------------------------------------------------------------------------------")
    print("|                                    Catálogo de Filmes Atualizado                                                                 |")
    print("----------------------------------------------------------------------------------------------------------------------------------")

    for codigo, filme in catalogo.items():
        titulo = filme['titulo'][:20] + '...' if len(filme['titulo']) > 20 else filme['titulo']
        genero = str(filme['genero'])[:20] + '...' if len(str(filme['genero'])) > 20 else str(filme['genero'])
        print(f"| {codigo:<7} | {titulo:<25} | Ano: {filme['ano']:<10} | Gênero: {genero:<24} | Avaliação: {filme['avaliacao']:<6} | Diária: R$ {filme['diaria']:<6} |")
        if filme['genero']:
            generos.extend(filme['genero'].split(','))

    print("------------------------------------------------------------------------------------------------------------------------------------")

    while True:
        codigo_filme = input("Selecione o código do filme para locação (ou 0 para encerrar): ")
        if codigo_filme == "0":
            break
        elif codigo_filme.isdigit() and int(codigo_filme) in catalogo:
            codigo_filme = int(codigo_filme)
            print(f"Valor da diária por filme: R$ {catalogo[codigo_filme]['diaria']:.2f}")
            quantidade_dias = int(input("Digite a quantidade de dias para locação: "))
            filme = catalogo[codigo_filme]
            preco_locacao = filme["preco"] * quantidade_dias
            total_locacao += preco_locacao
            filmes_locados[codigo_filme] = filme
        else:
            print("Opção inválida. Tente novamente.")
            continue
            
    print('-' * 70)
    print("\nTotal a pagar: R$ {:.2f}".format(total_locacao))
    print("Locação encerrada. Obrigado por usar nosso serviço!")
    print("Por favor, retorne para avaliar os filmes locados. Sua opinião é importante para nós!")
    print('-' * 70)

    for codigo, filme in filmes_locados.items():
        avaliacao = input(f"De 0 a 10, qual a sua avaliação para '{filme['titulo']}'? ")
        while not avaliacao.isdigit() or int(avaliacao) < 0 or int(avaliacao) > 10:
            print("Avaliação inválida. A avaliação deve ser um número de 0 a 10.")
            avaliacao = input(f"De 0 a 10, qual a sua avaliação para '{filme['titulo']}'? ")
        avaliacoes[codigo] = int(avaliacao)

    print("\nAvaliações dos Filmes:")
    for codigo, nota in avaliacoes.items():
        filme = catalogo[codigo]
        print(f"{filme['titulo']} - Avaliação: {nota}")
        print('-' * 70)

    return total_locacao, avaliacoes, generos



if not df_title_basics.empty and not df_title_ratings.empty:
    catalogo = catalogo_filmes(df_title_basics, df_title_ratings)

if catalogo is not None:
    total_locacao, avaliacoes, generos = locacao_filmes(catalogo)
else:
    total_locacao = 0
    avaliacoes = {}
    generos = []


if generos:
    generos_counts = pd.Series(generos).value_counts()
    plt.figure(figsize=(14, 8))
    plt.pie(generos_counts, labels=generos_counts.index, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 12})
    plt.title('Distribuição dos Gêneros dos Filmes no Catálogo', fontsize=18)
    plt.axis('equal')
    plt.show()


if avaliacoes:
    avaliacoes_df = pd.DataFrame({
        'Código do Filme': list(avaliacoes.keys()),
        'Avaliação': list(avaliacoes.values())
    })
    avaliacoes_df['Título'] = avaliacoes_df['Código do Filme'].apply(lambda x: catalogo[x]['titulo'])
    avaliacoes_df['Gênero'] = avaliacoes_df['Código do Filme'].apply(lambda x: catalogo[x]['genero'])

    plt.figure(figsize=(14, 8))
    sns.barplot(x='Título', y='Avaliação', hue='Gênero', data=avaliacoes_df, palette='viridis', dodge=False)
    plt.title('Avaliações dos Filmes Locados', fontsize=18)
    plt.xlabel('Título do Filme', fontsize=14)
    plt.ylabel('Avaliação', fontsize=14)
    plt.xticks(rotation=45, fontsize=12, ha='right')
    plt.yticks(fontsize=12)
    plt.legend(title='Gênero', fontsize=12)
    for index, row in avaliacoes_df.iterrows():
        plt.text(index, row['Avaliação'] + 0.1, f"{row['Avaliação']}", ha='center', fontsize=12)
    plt.tight_layout()
    plt.show()
```
